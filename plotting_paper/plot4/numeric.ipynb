### Import main packages

```python
import numpy as np
import figurefirst as fifi
import matplotlib as mpl
from mpmath import *
import matplotlib.font_manager
```

### Initialize plotting parameters

```python
# Plotting parameters
basefs=9
mpl.rcParams["mathtext.fontset"]    = "cm"
mpl.rcParams['font.family']         = 'Arial'
mpl.rcParams["font.size"]           = basefs
mpl.rcParams["axes.labelsize"]      = basefs+1
mpl.rcParams["xtick.labelsize"]     = basefs
mpl.rcParams["ytick.labelsize"]     = basefs
mpl.rcParams["legend.fontsize"]     = basefs-2
mpl.rcParams["legend.edgecolor"]    = "black" ;
mpl.rcParams["legend.fancybox"]     = True
mpl.rcParams["legend.borderpad"]    = 0.5
mpl.rcParams["legend.handlelength"] = .75
mpl.rcParams["lines.linewidth"]     = 1
mpl.rcParams["lines.markersize"]    = 2
mpl.rcParams["text.usetex"]         = False
mpl.rcParams["text.latex.preamble"] = "\\usepackage{amsmath,amsthm}; \\usepackage{physics}";
```

```python
filename = "numeric_layout" 

data_name = "numeric.npy"
HS = np.load(data_name)[[1, 3, 5], 0, 0] #[estimator][meas][nmeas][sample]
n_meas = [np.arange(3, 91, 3, dtype= int), np.arange(3, 91, 3, dtype= int), np.arange(5, 91, 5, dtype= int), np.arange(9, 91, 9, dtype= int), np.arange(3, 91, 3, dtype= int), np.arange(3, 91, 3, dtype= int)]
n_sample = 4000

xticks = np.array([0, 25, 50, 75])

layout = fifi.svg_to_axes.FigureLayout(
                                       filename + ".svg",
                                       autogenlayers=True, 
                                       make_mplfigures=True, hide_layers=[]
                                      )

recon =['Bayesian mean', 'maximum likelihood', 'direct inversion']
recon_l =['BME', 'MLE', 'direct inv.']
meas = ['Bellstate meas.', 'parity checks','MUB', 'Pauli meas.', 'Haar rand.', 'sep. Haar rand.']

c_meas = ['#009E73', 'plum', '#0072B2', '#D55E00', '#56B4E9', '#E69F00']
markers = ['D', 'x', 's']

lims = [[0, 0.22], [0, 0.66], [0, 0.66]]
yticks = np.array([[0, 0.1, 0.2], [0, 0.3, 0.6], [0, 0.3, 0.6]])
for idx, idn in enumerate([0, 1, 2]):
    ax = layout.axes[("fig", str(idx+1))]
    for i in range(len(meas)):
        temp = HS[idn][i][:len(n_meas[i])]
        std = np.std(1 - temp, axis=1) / np.sqrt(n_sample)
        #ax["axis"].errorbar(n_meas[i], np.average(temp, axis=1), yerr= std, c= c_meas[i], ls= "-", marker= markers[idn], alpha=1, zorder= 1, label= meas[i])
        if idx == 0: ax["axis"].plot(n_meas[i], np.average(temp, axis=1), c= c_meas[i], ls= "", marker= markers[idn], alpha=1, zorder= 3, label= meas[i])
        else: ax["axis"].plot(n_meas[i], np.average(temp, axis=1), c= c_meas[i], ls= "", marker= markers[idn], alpha=1, zorder= 3)

        
    ax["axis"].set_title(recon[idn])
    ax["axis"].set_xlim(0, 92)
    ax["axis"].set_ylim(*lims[idn])
    ax["axis"].set_xticks(xticks)
    ax["axis"].set_yticks(yticks[idn])
    ax["axis"].grid()
    ax["axis"].set_xlabel(r'$N$')
    
    if idn == 2:
        x = np.linspace(0.7, 91.3, 1000)

        ax["axis"].plot(x, 3/(5*x), c= c_meas[0], ls= '-', dash_capstyle= 'round', label= r'$3/(5N)$', alpha =0.5, zorder= 1)
        ax["axis"].plot(x, 3/(5*x), c= 'white', ls= '-', dash_capstyle= 'round', alpha =0.1, zorder= 2)

        ax["axis"].plot(x, 9/(5*x), c= c_meas[1], ls= '-', dash_capstyle= 'round', label= r'$9/(5N)$', alpha =0.5, zorder= 1)
        ax["axis"].plot(x, 9/(5*x), c= 'white', ls= '-', dash_capstyle= 'round', alpha =0.1, zorder= 2)

        ax["axis"].plot(x, 3/x, c= c_meas[2], ls= '-', dash_capstyle= 'round', label= r'$3/N$', alpha =0.5, zorder= 1)
        ax["axis"].plot(x, 3/x, c= 'white', ls= '-', dash_capstyle= 'round', alpha =0.1, zorder= 2)

        ax["axis"].plot(x, 27/(5*x), c= c_meas[3], ls= '-', dash_capstyle= 'round', label= r'$27/(5N)$', alpha =0.5, zorder= 1)
        ax["axis"].plot(x, 27/(5*x), c= 'white', ls= '-', dash_capstyle= 'round', alpha =0.1, zorder= 2)
    
    if idx != 1: 
        ax["axis"].legend()
    if idx == 0:
        ax["axis"].set_ylabel(r'average risk')

lims = [[0, 0.22], [0, 0.66], [0, 0.66], [0, 0.66]]
yticks = np.array([[0, 0.1, 0.2], [0, 0.3, 0.6], [0, 0.3, 0.6], [0, 0.3, 0.6]])
factors = [3/5, 9/5, 3, 27/5]
for i in range(4):
    ax = layout.axes[("fig", str(i+4))]
    x = np.linspace(0.7, 91.3, 1000)
    ax["axis"].plot(x, factors[i]/x, c= c_meas[i], ls= '-', dash_capstyle= 'round', alpha =0.5, zorder= 1)
    ax["axis"].plot(x, factors[i]/x, c= 'white', ls= '-', dash_capstyle= 'round', alpha =0.1, zorder= 2)
        
    for idn in range(3):
        temp = HS[idn][i][:len(n_meas[i])]
        std = np.std(1 - temp, axis=1) / np.sqrt(n_sample)
        #ax["axis"].errorbar(n_meas[i], np.average(temp, axis=1), yerr= std, c= c_meas[i], ls= "-", marker= markers[idn], alpha=1, zorder= 1, label= recon[idn])
        ax["axis"].plot(n_meas[i], np.average(temp, axis=1), c= c_meas[i], ls= "", marker= markers[idn], alpha=1, zorder= 1, label= recon_l[idn])

    
    ax["axis"].set_title(meas[i])
    ax["axis"].set_xlim(0, 92)
    ax["axis"].set_ylim(*lims[i])
    ax["axis"].set_xticks(xticks)
    ax["axis"].set_yticks(yticks[i])
    ax["axis"].grid()
    ax["axis"].set_xlabel(r'$N$')
    if i == 0: 
        ax["axis"].legend()
    if i == 0:
        ax["axis"].set_ylabel(r'average risk')
    
tmp = layout.save("numeric.svg")
```
